```python
import os
from imutils import paths
import cv2
from matplotlib import pyplot as plt
```

```python
def convert_image_to_greyscale(image):
    
    image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    return image
```

```python
def crop_image(image):
    
    height = image.shape[1]
    top_half_height = height / 2
    top_half_height = int(top_half_height)

    image = image[0:top_half_height, 0:]
    return image
```

```python
def manipulate_images():
    
    document_types = ['1099', 'applications', 'leases']
    for document_type in document_types:
        
        print('***** Manipulating ', document_type, ' documents. *****')
        
        document_type_directory = str.format('{}/{}', directory_where_original_images_live, document_type)
        image_paths = sorted(list(paths.list_images(document_type_directory)))
        
        i = 1
        for image_path in image_paths:
            
            # Load the image
            image_path = image_path.replace('\\', '')
            print('Loading: ', image_path)
            image = cv2.imread(image_path)
            
            # Convert the image to grayscale
            image = convert_image_to_greyscale(image)
            
            # Crop the image
            image = crop_image(image)
            
            # Save the image
            test_image_partial_path = str.format('{}/{}/Test', directory_where_manipulated_images_live, document_type)
            test_image_path = str.format('{}_{}.jpg', test_image_partial_path, i)
            i = i+1
            print('Saving: ', test_image_path)
            plt.imsave(test_image_path, image, cmap=plt.cm.binary_r)
```

```python
directory_where_original_images_live = '/Users/cpoole/Developer/Jupyter Notebooks/Practice Week/Fall 2018/Documents/Data/forms'
directory_where_manipulated_images_live = '/Users/cpoole/Developer/Jupyter Notebooks/Practice Week/Fall 2018/Documents/Data/Test/forms'

manipulate_images()
```
