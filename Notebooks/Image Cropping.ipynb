```python
import cv2
from matplotlib import pyplot as plt
```

## Load the image

```python
image = cv2.imread('1099.jpg')
```

```python
plt.imshow(image)
```

```python
print(image.shape)
```

## Convert the image to grayscale

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image, cmap=plt.cm.binary_r)
```

```python
print(image.shape)
```

## Crop the image so that only the top portion is kept

```python
image = image[0:600, 0:]
```

```python
plt.imshow(image, cmap=plt.cm.binary_r)
```

```python
print(image.shape)
```

## Crop the image so that only the top right portion is kept

```python
width = image.shape[0]
top_right_width = width / 2
top_right_width = int(top_right_width)
print('New width: ', top_right_width)

height = image.shape[1]
top_right_height = height /2
top_right_height = int(top_right_height)
print('New height: ', top_right_height)
```

```python
image = image[0:top_right_height, top_right_width:]
```

```python
plt.imshow(image, cmap=plt.cm.binary_r)
```

```python
print(image.shape)
```
